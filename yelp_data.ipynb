```python
# Import dependencies

import requests
import pandas as pd
import json
import numpy as np
from config import api_key
from pprint import pprint
from urllib.parse import quote
import matplotlib.pyplot as plt

```

```python
# Yelp wrapper code for pulling in food truck data from Yelp searches

def restaurant_response(set_num): 
# This function launches the request for all food truck endpoints in Los Angeles, CA.

    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {
        'Authorization': 'Bearer {}'.format(api_key),
    }
    url_params = { #parameters passed to the API
    "term": 'food trucks',
    "location": 90007,
    "state": "California",
    'offset': offset_num, # We are going to iterate the offset
     "limit":50 # Maximum return of results per request (ref: API documentation).
     }

    response = requests.get(url, headers=headers, params=url_params)
    response = response.json() # Returns a JSON.
    return response

for offset_num in np.arange(50,100,50) : 
    try:
        output_json = restaurant_response(offset_num) # Executing the function defined above.
        print(offset_num) # Making sure each offset iteration is running
        pprint(output_json,indent=4) # If you wanna check the JSON for each iteration
        if offset_num == 50:
            df_first = pd.DataFrame.from_dict(output_json['businesses'])

        else:
            df2 = pd.DataFrame.from_dict(output_json['businesses'])

            df_first = pd.concat([df_first, df2])
            
# The conditional statement above is so that I can append my results into a single dataframe, to 
# save into a single csv document.
    except AttributeError:
        print("error at ", offset_num) # Helpful for debugging purposes

#         df_first.to_json("./output/output_data.json")

# Saves the food truck data to csv file                        
            
df_first.to_csv("./output_data.csv", index = False)

# List for cities to search

search_cities = ['Los Angeles', 'Long Beach', 'Santa Clarita', 'Glendale', 'Lancaster', 'Palmdale', 'Pomona', 'Torrance', 'Pasadena', 'El Monte', 'Downey', 'Inglewood', 'West Covina', 'Norwalk', 'Burbank', 'Compton', 'South Gate', 'Carson', 'Santa Monica', 'Whittier', 'Hawthorne', 'Alhambra', 'Lakewood', 'Bellflower', 'Baldwin Park', 'Lynwood', 'Redondo Beach', 'Pico Rivera', 'Montebello', 'Monterey Park', 'Gardena', 'Huntington Park', 'Arcadia', 'Diamond Bar', 'Paramount', 'Rosemead', 'Glendora', 'Cerritos', 'La Mirada', 'Covina', 'Azusa', 'Bell Gardens', 'Rancho Palos Verdes', 'La Puente', 'San Gabriel', 'Culver City', 'Monrovia', 'Temple City', 'Bell', 'Manhattan Beach', 'Claremont', 'West Hollywood', 'Beverly Hills', 'San Dimas', 'Lawndale', 'La Verne', 'Walnut', 'Maywood', 'South Pasadena']



```

```python
df_food_trucks = pd.DataFrame([dict(name=d["name"],
                                    address=d["location"]["address1"],
                                   zipcode=d["location"]["zip_code"],
                                   rating=d["rating"],
                                   review_count=d["review_count"]) for d in restaurant_response["businesses"]])
```

```python
# Build a scatter plot comparing zip code to review count

fig, ax = plt.subplots(figsize=(20,5))
ax.scatter(df_food_trucks["zipcode"], df_food_trucks["review_count"])
ax.set(xlabel="zipcode", ylabel="review_count")

```

```python
df_food_trucks["zipcode"].value_counts()
```

```python
df_grouped = df_food_trucks.groupby(["zipcode"])

review_counts_by_zip = df_grouped["review_count"].sum()

df_review_counts_by_zip = review_counts_by_zip.reset_index()

df_review_counts_by_zip = df_review_counts_by_zip.sort_values("review_count", ascending = False)

df_review_counts_by_zip

```

```python
fig, ax = plt.subplots(figsize = (15,5))
ax.bar(df_review_counts_by_zip["zipcode"], df_review_counts_by_zip["review_count"], align = 'center')

ax.set(xticks=list(df_review_counts_by_zip["zipcode"]),
      xlabel = "Zipcodes",
      ylabel= "Total Number of Reviews",
      title = "Yelp Food Truck Review Counts by LA Zip Codes")

fig.tight_layout()
```
